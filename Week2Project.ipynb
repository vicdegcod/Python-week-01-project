```python
import random

number = random.randint(1 , 100)

print("The guess is:",number)

print("\nChoose a difficulty level:")
print("1. Easy (10 guesses)")
print("2. Medium (7 guesses)")
print("3. Hard (5 guesses)")


choice = input("Enter your choice (1/2/3): ")

if choice == "1":
    attempts = 10
elif choice == "2":
    attempts = 7
elif choice == "3":
    attempts = 5
else:
    print("Invalid choice! Defaulting to Easy level.")
    attempts = 10

remaining = attempts

# Game loop
while remaining > 0:
    try:
        guess = int(input(f"\nEnter your guess (Remaining guesses: {remaining}): "))

        if guess == number:
            print(f"Congratulations! You guessed the number {number} correctly!")
            print(f"You used {attempts - remaining + 1} attempt(s).")
            break
        elif guess < number:
            print("Too Low!")
        else:
            print("Too High!")

        remaining -= 1

        if remaining > 0:
            print(f"You have {remaining} guess(es) left.")

    except ValueError:
        print("Please enter a valid number.")

# If the player runs out of attempts
if remaining == 0:
    print(f"\nGame Over! You've used all your guesses.")
    print(f"The correct number was {number}.")
```
