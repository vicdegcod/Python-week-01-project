CLI Quiz Game Application
Build a Command-Line Based Quiz Application

```python

questions = [
    {
        "quiz": "Where is zindua located",
        "A": "Parklands",
        "B": "Westlands",
        "C": "Buruburu",
        "D": "CBD",
        "answer": "B",

    },
    {
        "quiz": "What is the learning mode at zindua",
        "A": "2 physical and 3 online classes",
        "B": "all physical classes",
        "C": "all online classes",
        "D": "2 online and 3 physical classes ",
        "answer": "A",
    },
    {
        "quiz": "One course offered at zindua",
        "A": "Architecture",
        "B": "Data science",
        "C": "Economics",
        "D": "Commerce",
        "answer": "B", 
    }
]

# Initialize score
score = 0

# Loop through questions
for i, q in enumerate(questions, start=1):
    print(f"\nQuestion {i}: {q['quiz']}")
    print(f"A. {q['A']}")
    print(f"B. {q['B']}")
    print(f"C. {q['C']}")
    print(f"D. {q['D']}")

    user_answer = input("Enter your answer (A, B, C, or D): ").upper()

    # Check answer using if statement
    if user_answer == q["answer"]:
        print("Correct!")
        score += 1
    else:
        print(f"Wrong! The correct answer is {q['answer']}.")

# Display final score
print("\nQuiz Completed!")
print(f"Your total score is: {score}/{len(questions)}")

# Provide feedback
if score == len(questions):
    print("Excellent!")
elif score >= 3:
    print("Good job!")
else:
    print("Try again!")
```
